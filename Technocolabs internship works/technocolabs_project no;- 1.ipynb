```python
# PROJECT 1,PART 1:-
# LINE CHARTS 
import pandas as pd 
df=pd.read_csv('unrate.csv')
df.head(5)
```

```python
df.date=pd.to_datetime(df.DATE)
df.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.DATE.head(12),df.VALUE.head(12))
plt.xticks(rotation=90)
plt.xlabel('month')
plt.ylabel('unemployment_rate')
plt.title("Monthly unemployment rate trends,1948")
plt.show()
```

```python
# PROJECT 1,PART 2:-
# MULTIPLE PLOTS.
```

```python
fig=plt.figure(figsize=(14,7))
fig_1948=fig.add_subplot(2,1,1)
fig_1949=fig.add_subplot(2,1,2)
fig_1948.plot(df[0:12]['DATE'],df[0:12]['VALUE'])
fig_1949.plot(df[12:24]['DATE'],df[12:24]['VALUE'])
fig_1948.set_title('Monthly Unemployment rate,1948')
fig_1949.set_title('Monthly unemployment rate,1949')
plt.show()
```

```python
import pandas as pd 
df=pd.read_csv('unrate.csv')



fig_1948_1952=plt.figure(figsize=(12,14))

for i in range(1953 - 1948):
    ax=fig_1948_1952.add_subplot(1953-1948,1,i+1)
    ax.plot(df[(i*12):(i+1)*12]['DATE'],df[(i*12):(i+1)*12]['VALUE'])
    ax.set_title('monthly unemployment rate,19'+str(48+i))
    plt.show()
```

```python

fig2_1948_1952=plt.figure(figsize=(20,6))
for i in range(1953-1948):
    plt.plot(df[(i*20):(i+1)*20]['DATE'],df[(i*20):(i+1)*20]['VALUE'])
    
plt.title('monthly unemployment rate,1948-1952')
plt.xlabel('month,integer')
plt.ylabel('unemployment rate,percentage')
plt.legend(['1948','1949','1950','1951','1952'],loc='upper left')
plt.show()
```

```python
import pandas as pd 
df=pd.read_csv('fandango_scores.csv')[["FILM","RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]]
```

```python
import matplotlib.pyplot as plt 
from numpy import arange
num_cols=['RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
bar_heights= df[num_cols].iloc[0].values
bar_positions= arange(5) + 0.75

fig,ax= plt.subplots()
ax.bar(bar_positions,bar_heights,0.5)

plt.xlabel("fandango")
plt.ylabel("rotten tomato")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
df=pd.read_csv('fandango_scores.csv')[["FILM","RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]]

num_cols = ['RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
bar_heights = df[num_cols].iloc[0].values
bar_positions= arange(5) + 0.75
tick_positions = range(1,6)

fig, ax= plt.subplots()
ax.bar(bar_positions,bar_heights,0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols,rotation = 90)

plt.xlabel("rating source")
plt.ylabel("average rating")
plt.title("average user rating for avengers: AGE OF ULTRON (2015)")
plt.show()
```

```python
# PROJECT 1, PART 4:-
# HISTOGRAM & BOX PLOTS.

import pandas as pd 
df=pd.read_csv('fandango_scores.csv')[["FILM","RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]]
```
